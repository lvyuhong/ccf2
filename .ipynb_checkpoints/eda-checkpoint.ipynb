```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gc
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.sans-serif']=['SimHei']   # 用黑体显示中文
plt.rcParams['axes.unicode_minus']=False     # 正常显示负号

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

# load data

```python
path = 'input/'
train_sales_data = pd.read_csv(path + '/Train/train_sales_data.csv')
train_search_data = pd.read_csv(path + '/Train/train_search_data.csv')
train_user_reply_data = pd.read_csv(path + '/Train/train_user_reply_data.csv')

test = pd.read_csv(path + '/evaluation_public.csv')
```

```python
test = test.merge(train_sales_data[['model','bodyType']],'left',on=['model'])

data = pd.concat([train_sales_data, test], ignore_index=True)
data = data.merge(train_search_data, 'left', on=['province', 'adcode', 'model', 'regYear', 'regMonth'])
data = data.merge(train_user_reply_data, 'left', on=['model', 'regYear', 'regMonth'])

data['yearmonth'] = data['regYear']*100+data['regMonth']
```

## bodyType

```python
month = ['201601','201602','201603','201604','201605','201606','201607','201608','201609','201610','201611','201612',
         '201701','201702','201703','201704','201705','201706','201707','201708','201709','201710','201711','201712']
type = ['salesVolume','popularity','carCommentVolum','newsReplyVolum']
```

```python
## 车身销售情况
df2 = data[(data['yearmonth']<201800)].groupby(['bodyType','yearmonth'])[type].sum()
bodyType = data['bodyType'].unique()
```

```python

for t in type:
    plt.figure(num=1,figsize=(12,8))
    for m in bodyType:
        plt.plot(month,df2.loc[(m)][t])
    plt.grid(True)
    plt.legend(labels=bodyType,loc='best')
    plt.title(t)
    plt.xticks(rotation=90)
    plt.show()
```
