```python
import ccf2
import numpy as np
import pandas as pd
import gc
import warnings
warnings.filterwarnings('ignore')

pd.options.display.max_columns = None
pd.options.display.max_rows = None
path = 'input/' 
```

```python
data = ccf2.loaddata(path)
```

```python
data.head()
```

```python
train_idx = (data['mt'] <= 24)

# valid_idx = (data['mt'].between(21, 24))

test_idx = (data['mt'] > 24)
```

```python
train = data[train_idx]
test = data[test_idx]
```

```python
test.head()
```

```python
submiss = pd.DataFrame()
# submiss['forecastVolum'] = train[train['regMonth']<=4].groupby(['adcode','model','regMonth'])['label'].mean().astype(int)
submiss['forecastVolum'] = data[data['mt'].between(13, 16)].groupby(['adcode','model','regMonth'])['label'].mean().astype(int)
submiss.reset_index(inplace=True)
```

```python
submiss.head()
```

```python
sub = test.merge(submiss,'left',on=['adcode','model','regMonth'])
```

```python
sub = sub[['id','forecastVolum']]
```

```python
sub.to_csv(path+'sub/secondModel.csv',index=False,header=True)
```
